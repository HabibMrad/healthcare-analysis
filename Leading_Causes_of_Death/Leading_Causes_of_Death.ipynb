```python
# This is a data exploration practice run on a dataset of
# leading causes of death in the US, using pandas, seaborn, and matplotlib.

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

data = pd.read_csv("NCHS_-_Leading_Causes_of_Death__United_States.csv")
simplified = data.drop(['113 Cause Name', 'Age-adjusted Death Rate'], axis=1)
```

```python
#data
```

```python
# Subsetting the data into deaths from kidney disease in Vermont
vermont_filter = simplified['State'] == 'Vermont'
vermont = simplified[vermont_filter]
vermont_kidney = vermont[vermont['Cause Name'] == 'Kidney disease']
vermont_kidney.head()
```

```python
# Plotting Deaths per year in Vermont due to Kidney disease. 
# Interesting downward trend. Need to fix x-ticks

sns.lmplot(x='Year', y='Deaths', data=vermont_kidney)
plt.ylim(0, 100)
plt.show()
```

```python
# Subsetting the data into deaths from suicide in Utah
suicide_filter = simplified['Cause Name'] == 'Suicide'
suicide_df = simplified[suicide_filter]
suicide_df
```

```python
utah_suicide_df = suicide_df[suicide_df['State'] == 'Utah']
utah_suicide_df

for state in set(suicide_df['State']):
    filtered_df = suicide_df[suicide_df['State'] == state]
    my_plot = sns.lmplot(x='Year', y='Deaths', data=filtered_df)
    my_plot.fig.suptitle(state + ' Suicide Deaths per Year')
    plt.show()
```

```python
# Plotting trend of suicide deaths in utah
sns.lmplot(x='Year', y='Deaths', data=utah_suicide_df)
plt.ylim(200, 800)
plt.show()
```
