```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
bucket_link = 'https://test-bucket-daniel-888.s3-us-west-2.amazonaws.com/NCHS_-_Drug_Poisoning_Mortality_by_State__United_States.csv'
#drug_poisoning = pd.read_csv('NCHS_-_Drug_Poisoning_Mortality_by_State__United_States.csv', sep=',')
drug_poisoning = pd.read_csv(bucket_link, sep=',')
alabama_filter = drug_poisoning['State'] == 'Alabama'
```

```python
drug_poisoning_alabama = drug_poisoning[alabama_filter]
```

```python
sns.lmplot(x='Year', y='Deaths', data=drug_poisoning_alabama)
plt.show()
```
